```python
import pandas as pd
import numpy as np

class Person:
    # 생성자 메서드
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # 인스턴트 메서드
    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# 객체 생성
person1 = Person("Alice", 30)
Person2 = Person("Bob", 25)


# 메서드 호출
person1.greet()
Person2.greet()

```
