```python
import numpy as np

# 배열생성
arr = np.array ([1,2,3,4,5])
print(f"Array:{arr}")
```

```python
# 다차원 배열생성
matrix = np.array([[1,2,3], [4,5,6]])
print(f"Matrix:\n{matrix}")
```

```python
# 배열 연산
arr_sum = np.sum(arr)
print(f"Sum of array:{arr_sum}")
```

```python
matrix_mean = np.mean(matrix, axis=0)
print(f"Mean of each column in marix:{matrix_mean}")
```

```python
#배열 슬라이싱
arr_slice  = arr[1:4]
print(f"Array slice : {arr_slice}")
```
