```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import matplotlib.pyplot as plt  # 한글 폰트 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'


df = sns.load_dataset('titanic')
print(df.head())
print(df.shape)
print(df.info())
```

```python
mis = df.isnull().sum()
total = len(df)

mis_percent = (mis / total) * 100
print(mis_percent)

print(pd.DataFrame({'결측수': mis, '결측비': mis_percent})[mis>0])
```

```python
# Fiure와 서브플롯 생성

# 생존여부 분포
fig, axes = plt.subplots(2, 2)
df['survived'].value_counts().rename({0:'사망', 1:'생존'}).plot(kind='bar', ax=axes[0,0],rot=0)
axes[0,0].set_title('생존 대 사망')

# 객실 등급별 생존율
df.groupby('pclass')['survived'].mean().plot(kind='bar', ax=axes[0,1],rot=0)
axes[0,1].set_title('객실 등급별 생존률')

# 성별 생존률
df.groupby('sex')['survived'].mean().plot(kind='bar', ax=axes[1,0],rot=0)
axes[1,0].set_title('성별 생존률')

# 나이 분포 (생존/사망)
# for s, color, label in [(0,'#e74c3c','사망'), (1,'#2ecc71','생존')]:
#     df[df['survived']==s]['age'].hist(
#         ax=axes[1,0], alpha=0.6, color=color, label=label, bins=20)

for s, color, label in [(0, '#e74c3c', '사망'), (1, '#2ecc71', '생존')]:
    df[df['survived']==s]['age'].hist(ax=axes[1,1], alpha=0.6, color=color, label=label, bins=20)
    axes[1,1].set_title('나이 분포 (생존/사망)')
    axes[1, 1].legend()

# 레이아웃 조정
plt.tight_layout()
plt.show()
```
