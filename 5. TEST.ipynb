```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats


df = sns.load_dataset('titanic')
print(df.head())
print(df.shape)
print(df.info())
```

```python
mis = df.isnull().sum()
total = len(df)

mis_percent = (mis / total) * 100
print(mis_percent)

print(pd.DataFrame({'결측수': mis, '결측비': mis_percent})[mis>0])
```
